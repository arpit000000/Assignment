Q1: Probability Mass Function (PMF) and Probability Density Function (PDF):

Probability Mass Function (PMF):
A Probability Mass Function (PMF) is used to describe the probability distribution of a discrete random variable. It gives the probability that the random variable takes on a specific value. In other words, it maps each possible value of the discrete random variable to its probability of occurrence.
Example: Consider rolling a fair six-sided die. The PMF of the random variable "X" (representing the outcome of the die) is given by:
PMF(X = 1) = 1/6
PMF(X = 2) = 1/6
PMF(X = 3) = 1/6
PMF(X = 4) = 1/6
PMF(X = 5) = 1/6
PMF(X = 6) = 1/6

Probability Density Function (PDF):
A Probability Density Function (PDF) is used to describe the probability distribution of a continuous random variable. It gives the relative likelihood of the continuous random variable taking on a particular value within a given range. The area under the PDF curve over a specific range corresponds to the probability of the variable falling within that range.
Example: The PDF of a standard normal distribution (mean = 0, standard deviation = 1) is given by the formula:
PDF(x) = (1 / √(2π)) * e^(-(x^2) / 2), where "x" is any value in the real number line.

Q2: Cumulative Density Function (CDF):

The Cumulative Density Function (CDF) is used to describe the cumulative probability distribution of a random variable. For a given value "x," the CDF gives the probability that the random variable is less than or equal to "x."

Example: The CDF of a standard normal distribution (mean = 0, standard deviation = 1) is given by the formula:
CDF(x) = ∫[-∞, x] (1 / √(2π)) * e^(-(t^2) / 2) dt, where the integral is taken from negative infinity to "x."

The importance of CDF is that it provides a convenient way to calculate probabilities for a specific range of values rather than individual points, which can be useful for various statistical analyses and hypothesis testing.

Q5: Bernoulli Distribution and Binomial Distribution:

Bernoulli Distribution:
The Bernoulli distribution is a discrete probability distribution representing a single random experiment with two possible outcomes: success (usually denoted as 1) and failure (usually denoted as 0). The distribution is defined by a single parameter "p," which represents the probability of success.
Example: Consider flipping a fair coin, where "success" is getting heads, and "failure" is getting tails. The probability of heads (success) is 0.5 (p = 0.5), and the probability of tails (failure) is also 0.5.

Binomial Distribution:
The binomial distribution is a discrete probability distribution representing the number of successes in a fixed number of independent Bernoulli trials. It is characterized by two parameters: "n" (the number of trials) and "p" (the probability of success in each trial).
Example: Suppose we flip a fair coin 5 times and want to know the probability of getting exactly 3 heads. Each flip is a Bernoulli trial with "p" = 0.5. The binomial distribution can be used to calculate the probability of getting 3 heads in 5 trials.

The key difference between Bernoulli and Binomial distributions is that Bernoulli deals with a single trial, while Binomial deals with multiple, independent trials and counts the number of successes.

Z-score (Standardized score) formula:
Z = (X - μ) / σ

Where:
X = the value we want to find the probability for (in this case, 60)
μ = the mean of the dataset (given as 50)
σ = the standard deviation of the dataset (given as 10)

Z = (60 - 50) / 10
Z = 1

However, we want the probability of the observation being greater than 60, which is the complement of the probability of being less than or equal to 60. Therefore:

Probability(X > 60) = 1 - Probability(X ≤ 60) = 1 - 0.8413 ≈ 0.1587

Q7: Uniform Distribution:

The uniform distribution is a continuous probability distribution where all values within a specified range are equally likely to occur. It is represented by a flat, constant probability density function (PDF) over the given interval. In other words, each value within the range has the same chance of being observed, and there are no preferential peaks or valleys in the distribution.

Example: Consider rolling a fair six-sided die. The outcomes (1, 2, 3, 4, 5, or 6) are all equally likely, and the probability of rolling any particular number is 1/6. This situation follows a discrete uniform distribution. In a continuous uniform distribution example, we can consider a random variable representing the time it takes to complete a certain task, where the task can be completed anywhere between 0 and 10 minutes, and each minute within that range is equally likely to be the completion time.

Q8: Z-Score and Its Importance:

The Z-score, also known as the standardized score, is a statistical measure that quantifies how many standard deviations a data point is away from the mean of a dataset. It is calculated using the formula:

Z = (X - μ) / σ

Where:
Z = Z-score
X = Data point
μ = Mean of the dataset
σ = Standard deviation of the dataset

Importance of Z-score:

Standardization: Z-scores standardize different datasets, allowing for easier comparison and analysis. It transforms the values of different datasets into a common scale, making it possible to compare observations from different distributions.

Outlier detection: Z-scores help identify outliers in a dataset. Data points with Z-scores significantly higher or lower than zero are considered outliers and might require further investigation.

Probability calculation: Z-scores are essential in finding probabilities associated with specific data points in a normal distribution. They can be used to determine the probability of a value occurring within a certain range or being greater or less than a particular value.

Q9: Central Limit Theorem (CLT):

The Central Limit Theorem (CLT) is a fundamental principle in statistics that states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the shape of the original population distribution. This theorem holds true under certain conditions.

Significance of the Central Limit Theorem:

Sampling theory: The CLT is crucial for inferential statistics because it allows statisticians to make inferences about the population mean using sample means. It enables us to draw conclusions about a population even when we only have access to a limited sample.

Normality assumption: It justifies the use of normal distribution-based statistical methods for sample means, even if the population distribution is not normal. This is particularly valuable since the normal distribution has well-known properties that facilitate hypothesis testing and confidence interval estimation.

Q10: Assumptions of the Central Limit Theorem:

For the Central Limit Theorem to hold, the following assumptions should be met:

Independence: The individual data points in the sample must be independent of each other.

Sufficient sample size: The sample size should be sufficiently large. While there is no fixed threshold, a common rule of thumb is that the sample size should be at least 30.

Finite variance: The population from which the samples are drawn should have a finite variance. In cases where the population variance is infinite, the CLT may not apply.

Identical distribution: Each sample in the dataset should be drawn from the same underlying population distribution.
