Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The ArithmeticError class is a built-in Python exception class that serves as the base class for a variety of mathematical error types. Some of the errors that are defined in this class include FloatingPointError, ZeroDivisionError, and OverflowError.

```python
try:
    a=3/0
    print(a)
except ZeroDivisionError as e:
    print(e)
    
```

```python
import sys

try:
    x = sys.maxsize
    y = x * x
except OverflowError:
    print("Error: Numeric overflow")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class is a built-in Python exception class that serves as the base class for a variety of indexing and key lookup errors. It is used to handle errors that occur when attempting to access an item in a collection (such as a list, tuple, or dictionary) using an invalid index or key.

```python
d={"a":1,"b":3}
try:
    a=d.get("c")
    print(a)
except KeyError:
    print("key error")
```

```python
l=[1,2]
try:
    a=l[3]
    print(a)
except IndexError as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a built-in Python exception that is raised when an imported module, package, or attribute cannot be found or loaded. 

```python
try:
    import abc
except ImportError as e:
    print(e)
```

Q6. List down some best practices for exception handling in python.

Use specific exception types: Instead of using a generic exception type like Exception, use a specific exception type that accurately describes the error that occurred. This makes it easier to understand the cause of the error and handle it appropriately.

Use try-except blocks sparingly: Try to limit the use of try-except blocks to the minimum necessary. Overuse of try-except blocks can make the code harder to read and understand.

Log exceptions: Logging exceptions can help with debugging and troubleshooting. Use a logging library to log exceptions along with relevant information such as the time of occurrence, the user, and the request.

Use finally block for cleanup: Use a finally block to perform cleanup operations such as closing files or releasing resources. The finally block will always be executed, even if an exception is raised or the program exits early.
